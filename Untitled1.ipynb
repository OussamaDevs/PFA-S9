```python
# ----------------------
# 1. Imports and Data Loading (FIXED)
# ----------------------
import pandas as pd
import joblib
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from xgboost import XGBRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('scaled_data.csv')

# Drop 'Date' column safely (no inplace=True)
data = data.drop(columns=['Date'])

# ----------------------
# 2. Correct Scaling (CRITICAL FIX)
# ----------------------
# Scale features and target SEPARATELY
feature_scaler = joblib.load("feature_scaler.pkl")
target_scaler = MinMaxScaler()

X = data.drop('Close', axis=1)
y = target_scaler.fit_transform(data['Close'].values.reshape(-1, 1))

# ----------------------
# 3. Time-Series Split (FIXED)
# ----------------------
train_size = int(len(data) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# ----------------------
# 4. LSTM Sequence Creation (WINDOW=15)
# ----------------------
def create_sequences(data, targets, window_size=15):
    X_seq, y_seq = [], []
    for i in range(len(data) - window_size):
        X_seq.append(data[i:i+window_size])
        y_seq.append(targets[i+window_size])
    return np.array(X_seq), np.array(y_seq)

X_train_seq, y_train_seq = create_sequences(X_train, y_train)
X_test_seq, y_test_seq = create_sequences(X_test, y_test)

# ----------------------
# 5. Simplified LSTM Model
# ----------------------
lstm_model = Sequential([
    LSTM(64, activation='tanh', input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])),
    Dense(1)
])

lstm_model.compile(optimizer='adam', loss='mse')

# Train with early stopping
history = lstm_model.fit(
    X_train_seq, 
    y_train_seq,
    epochs=50,
    batch_size=32,
    validation_data=(X_test_seq, y_test_seq),
    verbose=1
)

# ----------------------
# 6. Feature Engineering (FIXED SCALING)
# ----------------------
# Get LSTM predictions (SCALED SPACE)
train_lstm_pred = lstm_model.predict(X_train_seq)
test_lstm_pred = lstm_model.predict(X_test_seq)

# Inverse-transform predictions
train_lstm_unscaled = target_scaler.inverse_transform(train_lstm_pred)
test_lstm_unscaled = target_scaler.inverse_transform(test_lstm_pred)

# Inverse-transform original features for XGBoost
X_train_unscaled = feature_scaler.inverse_transform(X_train)
X_test_unscaled = feature_scaler.inverse_transform(X_test)

# Align data
X_train_xgb = X_train_unscaled[15:train_size]
X_test_xgb = X_test_unscaled[15:]

# Combine features with LSTM predictions
X_train_xgb = np.hstack([X_train_xgb, train_lstm_unscaled])
X_test_xgb = np.hstack([X_test_xgb, test_lstm_unscaled])

# Get true target values (UNSCALED)
y_train_xgb = target_scaler.inverse_transform(y_train[15:train_size])
y_test_xgb = target_scaler.inverse_transform(y_test[15:])

# ----------------------
# 7. XGBoost Training (TUNED)
# ----------------------
xgb_model = XGBRegressor(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=4,
    subsample=0.8,
    random_state=42
)

xgb_model.fit(X_train_xgb, y_train_xgb)

# ----------------------
# 8. Evaluation (PROPER UNSCALING)
# ----------------------
y_pred = xgb_model.predict(X_test_xgb)

mse = mean_squared_error(y_test_xgb, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test_xgb, y_pred)
r2 = r2_score(y_test_xgb, y_pred)

print(f"\nFinal Model Performance:")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"R²: {r2:.4f}")

# ----------------------
# 9. Visualization (FIXED)
# ----------------------
plt.figure(figsize=(15, 6))
plt.plot(y_test_xgb, label='Actual', color='blue')
plt.plot(y_pred, label='Predicted', color='orange', linestyle='--')
plt.title('Bitcoin Price Prediction (Fixed Hybrid Model)')
plt.xlabel('Time Steps')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```
